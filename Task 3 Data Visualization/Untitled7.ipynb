```python
import pandas as pd

df = pd.read_csv('all_books.csv')
```

```python
# Question:
# Which book rating appears most frequently in the dataset?

plt.figure(figsize=(8,5))

sns.countplot(x='Rating', data=df)

plt.title('Distribution of Book Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Books')

plt.show()
```

Insight:
This chart shows the frequency of each rating category and helps identify the most common ratings among books.

```python
# Question:
# How are book prices distributed across the dataset?

plt.figure(figsize=(10,6))

counts, bins, patches = plt.hist(df['Price'], bins=20)

# Add labels on bars
for count, patch in zip(counts, patches):
    plt.text(
        patch.get_x() + patch.get_width()/2,
        count,
        int(count),
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.title('Distribution of Book Prices')
plt.xlabel('Price (£)')
plt.ylabel('Number of Books')

plt.show()
```

Insight:
The histogram reveals the spread of book prices and highlights the most common price ranges.

```python
# Question:
# Do highly rated books tend to have higher prices?

plt.figure(figsize=(8,5))

sns.boxplot(x='Rating', y='Price', data=df)

plt.title('Price Distribution by Rating')
plt.xlabel('Rating')
plt.ylabel('Price (£)')

plt.show()
```

Insight:
This visualization compares prices across rating categories and helps determine whether higher-rated books are generally more expensive.

```python
# Question:
# What is the availability status of books in the dataset?

plt.figure(figsize=(8,5))

counts = df['Availability'].value_counts()

ax = counts.plot(kind='bar')

plt.title('Book Availability Status')
plt.xlabel('Availability')
plt.ylabel('Number of Books')

for i, value in enumerate(counts):
    plt.text(i, value + 2, str(value), ha='center')

plt.show()
```

Insight:
This chart displays the stock availability of books and provides an overview of inventory distribution.

```python
# Question:
# Is the Price column in numeric format?

df['Price'] = df['Price'].astype(str)

df['Price'] = df['Price'].str.replace('Â£', '', regex=False)
df['Price'] = df['Price'].str.replace('£', '', regex=False)

df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

print(df['Price'].dtype)
```

```python
# Question:
# Which books have the highest prices?

top_books = df.nlargest(10, 'Price')

plt.figure(figsize=(10,6))

sns.barplot(
    data=top_books,
    x='Price',
    y='Title'
)

plt.title('Top 10 Most Expensive Books')

plt.show()
```

Insight:
This visualization highlights the most expensive books in the dataset and identifies premium-priced titles.

```python
# Question:
# What are the key metrics of the dataset?

print("Total Books:", len(df))
print("Average Price:", round(df['Price'].mean(),2))
print("Highest Price:", df['Price'].max())
print("Most Common Rating:", df['Rating'].mode()[0])
```
