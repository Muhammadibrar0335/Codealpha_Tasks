```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv("all_books.csv")

# Check first rows
print(df.head())
```

```python
# Question:
# What is the distribution of book prices?

plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=20, kde=True)

plt.title("Distribution of Book Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# Question:
# Which ratings occur most frequently?

plt.figure(figsize=(8,5))
sns.countplot(x='Rating', data=df)

plt.title("Book Ratings Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
# Question:
# Is there any relationship between book price and rating?

plt.figure(figsize=(8,5))
sns.boxplot(x='Rating', y='Price', data=df)

plt.title("Price vs Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
# Question:
# Are there any unusual or extreme price values?

plt.figure(figsize=(8,5))
sns.boxplot(y=df['Price'])

plt.title("Price Outliers")
plt.show()
```

```python
# Question:
# Do highly rated books tend to have higher prices?

average_price = df.groupby('Rating')['Price'].mean()

print(average_price)
```

```python
# Question:
# Is there any correlation among numerical variables?

numeric_df = df.select_dtypes(include=np.number)

sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```

```python
# Question:
# What potential data quality issues exist in the dataset?

print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```
